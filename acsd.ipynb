```python
(h,w,c) = img.shape
```

```python
matr = cv2.getRotationMatrix2D((img.shape[1]//2, img.shape[0]//2), 90, 1)
rot = cv2.warpAffine(img, matr, (img.shape[1], img.shape[0]))
plt.imshow(rot)
```

```python
flip = cv2.flip(img, 1) # horizontal flip 1 = horizontal, 0 = vertical, -1 = both
plt.imshow(flip) # horizontal flip
```

```python
trans = cv2.transpose(img)
plt.imshow(trans) # transpose
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
img_inv = cv2.bitwise_not(gray)
plt.imshow(img_inv, cmap='gray')
```

```python
img2 = cv2.GaussianBlur(img_inv, (25, 25), sigmaX=0, sigmaY=0)
plt.imshow(img2, cmap='gray')
```

```python
img_p = cv2.divide(gray, 255-img2, scale=255)
plt.imshow(img_p, cmap='gray')
```

```python
cv2.imwrite("output.jpg", img_p)
```
